```python
import numpy as np

#类：线性回归-梯度下降
class LinearRegressionGD:
    def __init__(self, lr=0.01, epochs=2000):
        self.lr = lr  #学习率
        self.epochs = epochs  #迭代次数
        self.weights = None  #权重
        self.bias = None  #偏置
        self.losses = []  #记录每一轮迭代损失值

    #训练模型 选择参数
    def fit(self, X, y):
        n_smaples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for epoch in range(self.epochs):
            y_pred = np.dot(X, self.weights) + self.bias  #预测值y帽

            #梯度与更新
            dw = (1/n_smaples) * np.dot(X.T, y_pred - y)  #权重的偏导数
            db = (1/n_smaples) * np.sum(y_pred - y)  #偏置的偏导数
            self.weights -= self.lr * dw  #更新权重参数
            self.bias -= self.lr * db  #更新偏置参数

            loss = np.mean((y_pred - y) ** 2)  #成本函数
            self.losses.append(loss)

            #每200轮打印一次loss，判断是否收敛
            if epoch % 200 == 0:
                print(f"epoch: {epoch}, loss: {loss:.4f}")

    #使用训练好的参数进行预测
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

    #评估训练效果
    def evaluate(self, X, y):
        y_pred = self.predict(X)
        mse = np.mean((y_pred - y) ** 2)  #均方误差
        rmse = np.sqrt(mse)  #均方根误差
        mae = np.mean(np.abs(y_pred - y))  #平均绝对误差
        return {'mse': mse, 'rmse': rmse, 'mae': mae}  #包装成一个字典
```
