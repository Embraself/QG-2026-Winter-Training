#### **5 数据容器及其操作**

##### **1 列表list**

**(1)格式**：**列表名 = [v1, v2 ...]**

(注：列表元素可以为任意类型数据，也可以是另一个列表(嵌套))

(注：无元素即空列表，格式：`list1 = []` 或 `list2 = list()`)

**(注：列表元素 可以重复，可以修改)**

(注：列表可以不写列表名，直接写字面量)

(注：列表可以使用下标索引，**且负数意味着从尾部开始，即索引-1对应最后一个元素**)

**(2)列表的方法/函数**

* **index()函数：**

  格式：**列表名1.index(某元素)**

  作用：查找返回指定元素的下标索引，无则返回ValueError

* **insert()函数：**

  格式：**列表名1.insert(索引，元素值)**

  作用：=在指定索引位置，插入指定元素

* **append()函数：**

  格式：**列表名1.append(元素)**

  作用：将指定元素追加到列表尾部

* **extend()函数：**

  格式：**列表名1.extend(数据容器1)**

  作用：将指定数据容器1的所有元素取出，依次追加到列表尾部

* **pop()函数**

  格式：**列表名1.pop(索引)**

  作用：取出并返回对应索引的元素

  (注：取出后，原列表更新，被取出元素删除)

* **remove()函数**

  格式：**列表名1.remove(元素)**

  作用：删除指定元素的第一个匹配项

* **clear()函数**

  格式：**列表名1.clear()**

  作用：清空列表所有元素

* **count()函数**

  格式：**列表名1.count(元素)**

  作用：统计并返回对应元素在列表中的出现总数

* **len()函数**

  格式：**len(列表名1)**

  作用：统计并返回数据容器1的元素总个数

  *(注：已经不是列表专属)*

* **关键字del**

  格式：**del 列表名1[索引]**

  作用：删除指定索引的元素

  *(注：已经不是列表专属)*

**(3)列表的推导式**

* 格式：[表达式  for  tmp_v  in  迭代对象  if  条件]

* 作用：以简洁的语法将一个可迭代对象的元素依次取出待用，每一次循环表达式都最终成为列表的一个元素，最终**返回**一个**新的列表**

  *(注：if条件的可选的)*

  *(注：本质是简化了for+append()的写法)*






##### **2 元组tuple**

**(1)格式：元组名 = (v1, v2 ...)**

(注：元组元素可以为任意类型，也可以为另一个元组(嵌套))

(注：无元素即空元组，格式：`tuple1 = ()` 或 `tuple1 = tuple()`)

**(注：元组元素 可以重复，只读不改)**

**(注：元组括号内至少出现一个逗号，以区别字符串)**

元组`tuple1 = ("hello",)`  字符串`str1 = ("hello")`  前者有逗号后者无逗号

**(注：元组嵌套列表，嵌套列表的内容可以修改)**

(注：下标索引同列表)

**(2)元组的方法/函数**

* **index()函数：**

  格式：**元组名1.index(元素)**

* **count()函数：**

  格式：**元组名1.count(元素)**

* **len()函数：**

  格式：**len(元组名1)**



##### **3 字符串string**

**(1)格式：字符串名 = ”内容“**

(注：下标索引访问同列表)

**(注：元素为字符型，只读不改)**

**(2)字符串的方法/函数**

* **index()函数：**

  格式：**字符串名1.index(元素)**

* **count()函数：**

  格式：**字符串名1.count(元素)**

* **len()函数：**

  格式：**len(字符串名1)**

* **replace()函数：**

  格式：**字符串名1.replace(字符串a, 字符串b)**

  作用：将原字符串1副本中的字符串a替换为字符串2，然后返回

  **(注：不修改字符串1本身而是得一个新字符串)**

  (注：字符串a为字符串1的子串或本身)

* **split()函数：**

  格式：**字符串名1.split(分隔字符, 最大拆分次数)**

  作用：按指定的分隔字符串，将原字符串1划分成多个新子字符串，划分次数不超过最大拆分次数，并存入**列表**中，返回该列表

  (注：原字符串1不改变，得到一个新列表对象)

  (注：分隔字符串是原字符串1的一个子串)

  例：

  ```python
  str1 = "abcX1defX1gh"
  list1 = str1.split("X1")
  print(f"{list1}") #输出 ["abc", "def", "gh"]
  ```

* **strip()函数：**

  格式：**字符串名1.strip(指定字符串)**

  作用：去除前后指定的字符或字符集，指定字符串默认是**空白字符**

  (注：指定字符或指定字符集是**逐个匹配，不是整体匹配**)

  ```python
  s = "$$#hello#world$$#"
  result = s.strip("$#")  # 只要是$或#，首尾都会被删
  print(result)  # 输出：hello#world
  ```



##### **4 集合set**

**(1)格式：集合名1 = {v1, v2, ...}**

(注：元素可以为任意类型，亦可以是另一个集合(嵌套))

(注：无元素即空集合，格式：`set1 = set()`，**没有`set1 = {}`**的空集合写法，后者实为空字典)

**(注：集合元素 不可重复，可以修改)**

**(注：集合无序，不支持下标索引访问)**

**(2)集合的方法/函数**

* **remove()函数：**

  格式：**集合名.remove(元素)**

* **clear()函数：**

  格式：**集合名1.clear()**

* **pop()函数：**

  格式：**集合名1.pop()**

  作用：从集合中**随机抽取**一个元素出来并返回，原集合被修改

  (注：集合不支持下标索引，所以pop()中没有索引参数)

* **add()函数：**

  格式：**集合名1.add(元素)**

  作用：将指定元素加入到集合中

* **difference()函数：**

  格式：**集合名1.difference(集合名2)**

  作用：返回一个集合1与集合2的**差集**，差集为集合1中有而集合2中无的集合

* **difference_update()函数：**

  格式：**集合名1.difference_update(集合名2)**

  作用：仅仅改变集合1，将集合1内的与集合2相同的元素全部删除，集合2不改变

  (注：**没有返回值**，不返回新集合1，原集合1直接被改变)

* **union()函数：**

  格式：**集合名1.union(集合名2)**

  作用：返回集合1与集合2组合得到的新集合，两个原集合不变



##### **5 字典dict**

**(1)格式：字典名1 = {k1: v1, k2: v2, ...}**

(注：每一个元素都是一对**键值对**，**其中key唯一且不可为字典类型**，**value可以重复且可为字典(嵌套)**)

(注：无元素即空字典，格式：**`dict1 = {}`**或`dict1 = dict()`，**注意前者不是空集合写法**)

(注：字典不支持下标索引访问，访问格式：**字典名1[key] 去访问对应value**)

**(2)字典的方法/函数**

* **新增元素(访问)：**

  格式：**字典名[新key] = 新value**

  作用：添加新元素(键值对)到字典中

* **更新元素(访问)：**

  格式：**字典名[旧key] = 新value**

* **pop()函数：**

  格式：**字典名1.pop(key)**

  作用：返回对应key的value，原字典被修改

* **clear()函数：**

  格式：**字典名.clear()**

* **len()函数：**

  格式：**len(字典名1)**

* **keys()函数：**

  格式：**字典名.keys()**

  作用：返回一个包含所有key值的**列表**

  ```python
  dict1 = {'a': 100, 'b': 50, 'c': 0}
  for key1 in dict1:
      print(f"{key1}")
  for key2 in dict1.keys():
      print(f"{key2}")
  # 二者输出一样，后者显示写法，完全等价
  ```

**7 序列与切片，解包**

**(1)序列：内容连续有序，支持下标索引的一类数据容器(如列表，元组，字符串...)**

**(2)切片：从序列中取出一个子序列的操作**

格式：**序列名1[起始下标: 结束下标: 步长]**

(注：起始/结束下标可以省略，默认从开头元素(包含)起始，到末尾元素(不包含)结束，区间**左闭右开**)

(注：步长为摘取元素的间隔大小，省略默认为1，为负数则从末尾开始摘取)

(注：返回值(数据容器)类型与被切片的序列类型一致)

**(3)解包：把一个可迭代对象里元素，按顺序取出并”拆开来“，赋值给等号左边的多个变量**

(注：等号左边的变量数目必须与迭代器内元素数目**相等**(特殊情况除外))

**8 常见报错异常类型**

1. **SyntaxError（语法错误）**

   * 含义：代码不符合py语法规则

   * 报错提示：`SyntaxError: invalid syntax`

2.  **NameError（名称错误）**

   * 含义：使用了未定义的变量 / 函数 / 类名
   * 报错提示：`NameError: name 'xx' is not defined`

3. **TypeError（类型错误）**

   * 含义：对不同类型的变量执行了不支持的操作

   * 报错提示：`TypeError: unsupported operand type(s) for +: 'int' and 'str'`

     ```python
      # 上述报错提示的反面例子：
         1 + "2"  # 整数和字符串不能相加
     [1,2].append("3", "4")  # append只能传1个参数，传了2个
     {"name": "Tom"}[0]  # 字典只能用键取值，不能用索引
     ```

4. **IndexError（索引错误）**

   * 含义：使用了超出序列（列表 / 字符串 / 元组）范围的索引
   * 报错提示：`IndexError: list index out of range` *(以 list 为例子)*

5. **KeyError（键错误）**

   * 含义：访问字典中不存在的键或键值笔误
   * 报错提示：`KeyError: 'xx'`

6. **AttributeError（属性错误）**

   * 含义：访问了对象不存在的属性 / 方法

   * 报错提示：`AttributeError: 'list' object has no attribute 'add'`

     ```python
     lst = [1,2,3]
     lst.add(4)  # 列表没有add方法（正确是append）
     s = "hello"
     print(s.length)  # 字符串没有length属性（正确是len(s)）
     ```

     *(注：split()函数是字符串方法，切忌用于 列表list)*

7. **ValueError（值错误）**

   * 含义：数据类型正确，但值不符合要求

   * 报错提示：`ValueError: invalid literal for int() with base 10: 'abc'`

     ```python
     int("abc")  # 字符串不能转成整数
     lst = [1,2,3]
     lst.remove(4)  # 列表里没有4这个值
     ```

8. **IndentationError（缩进错误）**

   * 含义：专门指缩进相关的语法错误

   * 报错提示：`IndentationError: expected an indented block`

     ```python
     if 1 > 0:
     print("yes")  # 没有缩进
     def func():
         print("hello")
       print("world")  # 缩进不一致（有的4空格，有的2空格）
     ```

     *(注：属于SyntaxError 的子类，Python 特有的)*

  



​                      





9. **NotImplementedError（方法未实现错误）**

   * 含义：强制子类必须复写某个方法，否则调用时报错

     *(注：一般搭配**raise关键字**使用，主动抛出错误)*

  

#### **8 自定义模块导入**

**1 含义：**自定义一个自己设计的模块，本质为一个py文件，并为外部使用

*(注：导入方法语法与py自带模块的一样，直接写文件名)*

*(注：自定义模块名的命名规则应该符合标识符命名规则)*

*(注：若多次导入不同模块的对应方法，有的模块出现了同名，则后导入的方法覆盖先导入的方法)*

*(注：导入过程中，被导入模块内全部代码会被执行一次)*



**2 ______main______变量与______name______变量：**

* 格式：if ______name______ == '______main______':
* 作用：在导入模块时，变量______name______保持不变，if下的代码无法执行，直接在本模块运行时，变量______name______转化为______main______变量。使if下的代码可以运行

*(注：两个变量均是py内部的自带变量)*



**3 ______all______变量：**

* 作用：导入一个内带______all______变量时，以“*”方式导入全部，则只有变量______all______中的元素可以导入

*(注：变量______all______本质为一个列表)*

```python
#模块mod1
__all__ = ["test_A"]
def test_A():
    print('A')
def test_B():
    print('B')
#from mod1 import * 时，只会导入test_A(),不会导入test_B()
```

  ​           方法体

*(注：类的属性即成员变量，类的行为即成员方法)*

*(注：在类的方法内部，欲想访问到类的成员变量，则必须使用"**self.成员变量名**")*

*(注：定义类的方法时必须写self，但定义后再使用时，可以不管self，也无需传值)*



**2 ______init______()构造方法：**

* 格式：#在类当中

  ​            def ______init______(self, 形参...):

  ​                     方法体

*(注：在创建类对象时，**______init______()方法会自动执行**)*

*(注：在创建类对象时，给类括号传参值，参值会被传到______init______方法去任其使用)*



**3 常用的内置方法（魔术方法）：**

* ______str______()方法（字符串方法）：

  * 格式：#在类当中

    ​          def ______str______(self, 形参...):

    ​                 方法体

    ​                 return “xxx”     #给类对象返回字符串“xxx”

  * 作用：**定义**类对象被print()，str()等函数调用时的**格式**，**返回一个字符串**



* ______lt______()方法（小于符号比较方法）：

  * 格式：#在类当中

    ​            def ______lt______(self, other):

    ​                   方法体

    ​                   return self.var < other.var     #定义小于比较符的使用规则

  * 作用：定义同类对象之间使用小于号（<）的规则，**返回bool值**

*(注：形参other是另一个同类对象)*



* ______le______()方法（小于等于符号比较方法）：

  * 格式：#在类当中

    ​           def ______le______(self, other):

    ​                  方法体

    ​                  return self.var <= other.var     #定义小于等于比较符的使用规则

  * 作用：定义同类对象之间使用小于等于号（<=）的规则，**返回bool值**

*(注：形参other是另一个同类对象)*

*(注：在使用______lt______()，______le______()方法时，一定要先判断**other**的类型**是否为同类型**，切忌不同类相比)*



* ______eq______()方法（等于符号比较方法）：

  * 格式：#在类当中

    ​            def ______eq______(self, other):

    ​                   方法体

    ​                   return self.var == other.var     #定义等于比较符的使用规则

  * 作用：定义同类对象之间使用等于号（==）的规则，**返回bool值**

*(注：若无______eq______()方法而直接比较两个同类对象，则比较的是二者的内存地址，一般都是False)*



**4 封装，继承和多态：**

**（1）封装：**

* 含义：给类对象的属性和方法“打包”，只暴露必要的外部接口，隐藏内部具体的实现

* 私有化：

  * 格式：在成员或方法命名时，**在开头加上“双下划线”**
  * 作用：类对象无法直接访问私有成员/方法，提供**仅供内部使用**的属性/方法

  *(注：类的成员或方法可以直接访问私有的成员/方法)*



**（2）继承：**

* 含义：一个类继承另一个类的属性和方法，实现代码的复用，同时允许**子类拓展复写**新的属性方法

* 格式1：（单继承）

  ​          class 子类名(父类名):

  ​                    类内容体

* 格式2：（多继承）

  ​          class 子类名(父类1, ... 父类n):

  ​                    类内容体

*(注：在多继承中，若出现不同父类但**同成员变量名/方法名**，则**默认从左到右**的优先级，先继承的保留后继承的被覆盖)*



* pass关键字：
  * 作用：在继承/书写方法体函数体时想**写空**，为符合语法完整性可以单写**pass**



* 复写与调用父类成员：

  * 复写含义：子类在继承父类的属性和方法后，对旧的属性方法**重新定义书写**

  * 调用父类成员：

    * 方式一：父类名.成员变量

      ​                父类名.成员方法(self)

    * 方式二（super()调用）：

      ​               super().成员变量

      ​               super().成员方法()



**（3）多态：**

* 含义：多种状态，即完成某行为时，使用不同对象会得到不同状态，**根据具体对象决定具体行为**

  *(注：以父类做定义声明，以子类做实际工作，用以获得同一行为不同状态)*

  

**5 类型注解：**

**（1）含义：**给自定义的变量，方法/函数补充书写时的类型解释，以方便自己编码和IDE等第三方工具

**（2）格式：**

* 变量类注1：变量: 类型 = 初始化

```python
var1: int = 100
```

* 变量类注2: 

  ​                    class 类名:

  ​                            pass

  ​                            var: 类名 = 类名()

  

* 容器类注：变量名: 容器类别[元素类别, ...] = 初始化

* 注释类注：... ...     #type: 类型

* 函数/方法及其返回值类注:

  ​                     def 函数方法名(形参: 类型, ...) -> 返回值类型:

  ​                                   函数方法体

  ```python
  def func(var1: int, var2: list) -> str:
      pass
      return xxx
  ```

  

*(注：类型注解是**可选的**，不是决定性的，即类型注解与实际不符也**不会运行报错**)*

**（3）Union类型：**

* 格式：Union[类型1, 类型2, ... 类型n]
* 作用：定义联合类型注解，即对应元素的类型为**Union中的任意一种**

*(注：需要导入**Union包**)*

```python
my_list: list[Union[int, str, float]] = []  #列表my_list中元素类型可为int,str,float
```
