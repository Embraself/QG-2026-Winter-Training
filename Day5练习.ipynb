```python
from math import pi     #从math包导入圆周率pi

class Shape:
    count: int = 0
    def __init__(self):
        Shape.count += 1
        self.__area: float = 0.0

    def _calc_area(self) -> float:
        raise NotImplementedError     #要求子类必须复写该方法

    def get_area(self) -> float:
        self.__area: float = self._calc_area()
        return self.__area


class Rectangle(Shape):
    def __init__(self, length: float, width: float):
        super().__init__()
        self.length = length
        self.width = width

    def _calc_area(self) -> float:     #复写父类方法_calc_area()
        return self.length * self.width


class Circle(Shape):
    def __init__(self, radius: float):
        super().__init__()
        self.radius = radius

    def _calc_area(self) -> float:     #复写父类方法_calc_area()
        return pi * (self.radius ** 2)


if __name__ == '__main__':
    circ1 = Circle(3.0)
    circ2 = Circle(4.0)
    rect1 = Rectangle(3.0, 4.0)
    rect2 = Rectangle(15.5, 20.5)

    print(circ1.get_area())
    print(circ2.get_area())
    print(rect1.get_area())
    print(rect2.get_area())
    print(Shape.count)

    print(rect1._calc_area())
```
