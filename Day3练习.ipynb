```python
str1 = " Agent:007_Bond; Coords:(40,74); Items:gun,money,gun; Mission:2025-RESCUE-X "
str2 = str1.replace(" ","")  #去除所有空格
print(str2)

list1 = str2.split(";")  #以；分割成列表
print(list1)

tmp_list = list1[2].strip("Items:").split(",") #去除"Items:"并以','分割成临时列表
print(tmp_list)

set1 = set(tmp_list)  #集合去重,得到集合 {'gun','money'}
print(set1)

slice_of_str = list1[3][8:] #得到"2025-RESCUE-X"
print(slice_of_str)

list_for_tuple = list1[1].split(":")[1].strip("()").split(",")  #得到列表 [40,74]
tuple1 = tuple(map(int, list_for_tuple))  #map()函数，得到元组 (40，74)
print(tuple1)

name_of_agent = list1[0][6:]  #获取特工ID
print(name_of_agent)

dict1 = {
    "特工ID": name_of_agent,
    "坐标": tuple1,
    "装备": set1,
    "核心任务": slice_of_str,
}  #规整成新字典
print(dict1)
```
