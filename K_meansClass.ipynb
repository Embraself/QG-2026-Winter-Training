```python
import numpy as np

class Kmeans:
    def __init__(self, k, max_iter):
        self.k = k  #聚类堆数
        self.max_iter = max_iter  #最大迭代次数
        self.centroids = None  #聚类中心（质心）
        self.lables = None  #每堆的标签


    def _assign_clusters(self, X):
        distances = np.linalg.norm(X[:, np.newaxis] - self.centroids, axis=2)  #二维数组，每行每个元素是每个样本与k个质心的欧式距离
        return np.argmin(distances, axis=1)  #返回每个样本分配到的簇标签


    def update_centroids(self, X):
        new_centroids = np.zeros((self.k, X.shape[1]))  #新的质心（集合）
        for i in range(self.k):
            cluster_points = X[self.lables == i]  #将属于第i个质心的所有样本取出
            if len(cluster_points) > 0:  #证明有元素被分配到这第i个的质心
                new_centroids[i] = np.mean(cluster_points, axis=0)  #新质心的每个特征的值都是在cluster_points中取的对应特征的均值
            else:
                new_centroids[i] = self.centroids[i]  #保存第i个质心不变
        return new_centroids


    def fit(self, X):
        n_samples = X.shape[0]  #样品总数（行）
        random_indices = np.random.choice(n_samples, self.k, replace=False)  #随机取k个质心的索引数
        self.centroids = X[random_indices]  #存储质心（三个中心点，每个是11维向量（红酒特征））

        for _ in range(self.max_iter):
            self.lables = self._assign_clusters(X)  #每个样本的标签号（表明分配到哪一堆）
            old_centroids = self.centroids.copy()  #保留旧质心，用于判断是否收敛
            self.centroids = self.update_centroids(X)  #更新质心

            #判断收敛
            if np.all(np.abs(old_centroids - self.centroids) < 1e-6):
                print(f"第{_+1}步收敛")
                break

    def predict(self, X):
        distances = np.linalg.norm(X[:, np.newaxis] - self.centroids, axis=2)
        return np.argmin(distances, axis=1)

```
