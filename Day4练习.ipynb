```python
num_of_list = [tmp**2 for tmp in range(1, 11)] #range() 区间左闭右开
print(num_of_list)

name_of_list = ["Tom", "Mike", "Jack"]
modified_list = list(map(lambda x: "QG_"+x, name_of_list))
print(modified_list)

raw_data = ["85", "92", "ERROR", "105", "78", "WARNING", "99", "120"]
for item in raw_data:
    try:
        if int(item) / 100 >= 0.80:
            print(int(item)/100)
            print("核心过载")
        elif int(item) / 100 < 0.80:
            print(int(item)/100)
            print("运转正常")
    except ValueError as e:
        print(f"元素{item}不合要求，报错：{e}")
```
