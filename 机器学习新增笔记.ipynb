## **机器学习：**

### **1 机器学习概念：**

* **概念**：使机器在**没有明确编程代码规则**的情况下，**自身通过学习**数据中的规律，总结出“规则”，从而对**新数据做出预测或判断**



### **2  深度学习概念：**（机器学习进阶）

1. **概念**：在机器学习基础上，使用**深度神经网络**（多层结构），以更好地学习数据中的复杂模式
2. **基础神经网络（三层）和数据集划分**：
   1. **基础神经网络：**
      * 输入层：接收原始数据
      * 隐藏层：对数据进行特征提取和加工
      * 输出层：输出最终结果
   2. **数据集划分：**
      * 训练集：给模型提供学习的数据，让其**学习并总结**其中规律，一般占比最大
      * 验证集：在训练过程中使用，用来**调整**模型的**超参数**和选择**模型结构**，防止过拟合
      * 测试集：训练完成后使用，用来评估模型在**全新、未见过的数据**上的真实表现，衡量**泛化能力**



### **3 ML基础知识：**

1. **神经网络：**

   1. **概念：**一种由大量相互连接的处理单元（神经元）分层组成的**计算模型**，它通过从数据中**学习权重**来模拟人脑的信息处理方式，从而实现模式识别、预测和决策等智能任务

      *(补：神经网络**架构**：由大量神经元连接，信息从输入层流入，经过隐藏层的层层处理，从输出层流出)*

   2. **神经元：**

      * 含义：神经网络的**最小计算单位**，接收**多个输入**，并对其进行**加权求和**，通过**激活函数**输出一个值（单个神经元向前传播）

        *(注：y=f(w1x1+w2x2+...+wnxn+b)，其中w为**权重**，x为不同**数据**，b为**偏置**，y即**激活函数**)*

   3. **向前传播：**（做预测）

      * 含义：数据从输入层进入，经过每一层，每一层的输出结果作为下一层的输入数据，最终从输出层输出的**过程**

   4. **反向传播：**（“学习”的核心算法）

      * 含义：三个步骤——过程
        * （一）网络做出预测后，使用**损失函数**计算**预测值**与**实际值**的差距（损失）
        * （二）反向传播会从**输出层**开始，把这个 “错误” 逐层**向前传递**，计算每个神经元**对这个错误的 “贡献”**，然后**调整**每个神经元的**权重**，让下一次预测更准确
        * （三）这个过程**反复迭代**，直到模型预测做得足够好

   5. **激活函数：**

      * 含义：给神经网络注入**“非线性性”**，使模型不等价于线性模型**以便学习更复杂的规律**
      * 常见激活函数：
        * **ReLu：**
        * **Sigmoid：**
        * **Tanh：**

   6. **损失函数：**

      * 含义：用以衡量**预测值**与**实际值**之间**差距**的函数，以指导“学习”方向而最终最小化这个**损失值**
      * 常见的损失函数：
        * **MSE：**（均方误差）
        * **Cross-Entropy：**（交叉熵）
        
      
   7. **拟合泛化与模型评估指标：**
   
      1. **拟合：**
   
         * **欠拟合：**训练集误差、验证集误差和训练集误差均很高，**偏差很大**
         * **正常拟合：**训练集误差**适中**，验证集误差接近，**忽略个别噪音**
         * **过拟合：**训练集误差极低，验证集误差**明显升高**，**吸收过多噪音**
   
      2. **泛化：**
   
         * 含义：模型在**从未见过**的数据上**作出正确预测**的能力（“泛化好”：即模型训练的最终目的）
   
         * **泛化误差：**泛化误差 = 偏差 + 方差 + 噪音
   
           (注：**偏差**（欠拟合主导）：预测值与真实值的系统性偏离；**方差**（过拟合主导）：模型对训练数据细微变化的**敏感程度**；**噪音**：数据本身固有的难以避免的误差)
   
      3. **模型评估指标：**
   
         * 含义：**量化**模型好坏的数字标尺，用来判断**拟合程度**和**泛化能力**
         * **预测类别：**（分类任务）
           * 准确率：
           * 精准率：
           * 召回率：
           * F1分数：
           * AUC-ROC：
         * **预测数值：**（回归任务）
           * MSE（均方误差）：
           * MAE（平均绝对误差）：
           * R^2（绝对系数）：
   
2. **三大学习范式：**（是否带有**“标签”**）

   1. **监督学习：**（自带“标签”）

      * 目的：学习从输入到输出的映射

      * 特点：数据自带**提示性“标签”**，意义明显

        *(例：明确告知机器，该图片是“猫 ”)*
   2. **无监督学习：**（完全无“标签”）

         * 目的：促使机器**自己寻找**数据的内在模式/**内在特质**/本质
         * 特点：数据**没有标签**，机器**自己寻找**学习，发现，总结数据中的结构和规律

   3. **半监督学习：**（少量“标签”+大量无“标签”）
      * 目的：利用**少量标签数据**和**大量无标签数据**来**提升模型性能**
         * 特点：**小部分**数据有”标签“，**大部分**数据无“标签”

3. **线性回归模型：**

   1. **线性回归模型：**

      * 作用：用一条直线（或超平面）来**拟合数据**，作出预测或提前判断

      * 表达式：
        $$
        f(x)=wx+b
        $$
        *(注：其中参数**w**为**权重**（斜率），用以表示特征对预测结果的影响程度和方向)*

        *(注：其中参数**b**为**偏置**（截距），用以表示当所有特征为 0 时的预测基准值)*

   2. **成本函数--均方误差（MSE）：**

      * 作用：衡量模型**预测值 **f(x) 和**真实值** y 之间的差距；数值越小，误差差距越小

      * 公式：
        $$
        J(w,b) = \frac{1}{2m} \sum_{i=1}^{m} \left( f(x^{(i)}) - y^{(i)} \right)^2
        $$
        
      * **J(w,b)图像：**三维网状“碗形”（随w，b变化）
      
   3. **线性回归梯度下降：**（“下山算法”）
   
      * **作用：**不断更新优化参数w和b**，使成本函数值取**尽可能小的值，使得线性回归模型拟合更优
   
        *(注：对于**线性回归的 MSE 成本函数**，它是一个凸函数，只有**一个全局最小值**，梯度下降**总能找到**。**但对于复杂的深度神经网络，成本函数可能有**多个局部最小值**，梯度下降可能会因在某个局部最小值，**梯度导数为零**而**更新不了**两个参数，**停止更新**)*
      
      * **参数w与b的更新：**
        $$
       w = w - \alpha \frac{\partial J}{\partial w}
        $$
      
        $$
       \frac{\partial J}{\partial w} = \frac{1}{m} \sum_{i=1}^{m} \left( f_{w,b}(x^{(i)}) - y^{(i)} \right) \cdot x^{(i)}
        $$
   
        $$
   b = b - \alpha \frac{\partial J}{\partial b}
        $$
   
        $$
   \frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} \left( f_{w,b}(x^{(i)}) - y^{(i)} \right)
        $$
      
        *(注：x与y在式子对w求导时均为**常数**，设整个括号为**u**（关于w的函数）进行链式求导)*
        
        *(注：**梯度导数**：本质是J(w)对w的切线**斜率**，控制梯度下降的**步长**和下降的**方向**（符号正负），帮助 J(w ,b)取向 min值，图像越陡下降越快)*
        
        *(注：**学习率**：控制梯度下降的**步长**，取值范围(0,1))*
      
        *(注：公式**等号**，即有编程语言中**“赋值 ”**的意义，也有数学运算中**“断言相等 ”**的意义)*
   
        *(注：梯度下降要求**同时更新**参数**w**和**b**的值，代码中两个**赋值语句**应写于两个**更新语句**之后，否则其中一条更新语句将被前面的赋值语句**干扰**从而出错)*
   
        ```python
        # “同时更新w和b” 正确顺序：两更新语句+两赋值语句
        tmp_w = w - α(∂/∂J)J(w)
        tmp_b = b - α(∂/∂J)J(b)
        w = tmp_w
        b = tmp_b
        ```
        
        *(注：随着下降不断进行，梯度导数越来越小，w和 b的变化将**越来越不显著**，即**收敛**)*
        
      * **学习率α：**（梯度下降实现的效率）
      
        * 作用：更新选择参数**w**和**b**的**调整幅度**，每次迭代模型都根据当前数据误差来调整参数**w**和**b**以降低**J(w,b)**的值，拟合更优线性回归直线
        * 特点：（1）当**α**取值**过大**，即调整幅度过大，易在最优**J(w,b)**值附件**震荡**，靠近min值**缓慢**，甚至**发散**而**远离min值**。（2）当**α**取值**过小**，即调整幅度过小，**收敛极慢**，容易**卡在局部最优**，靠近min值**缓慢**，训练效率极低
        * **取值原则：**
          1. 训练初期用较大的学习率快速接近最优解，后期用较小的学习率精细调整
          2. **观察成本曲线**：如果曲线震荡或上升，说明学习率太大；如果几乎不动，说明太小
   
4. **优化线性回归模型：**
   
   1. **多特征与向量化：**
   
      1. **多元线性回归模型：**（多特征）
         $$
         f_{\vec{w},b}(x) = w_1 x_1 + w_2 x_2 + \dots + w_n x_n + b
         $$
         其中x1，x2...xn均是特征之一
   
      2. **向量化：**
   
         * 参数w向量化：
           $$
           \vec{w} = [w_1, w_2, \dots, w_n]
           $$
           **权重向量**，每个元素对应一个特征的重要程度
           
         * 特征x向量化：
           $$
           \vec{x} = [x_1, x_2, \dots, x_n]
           $$
           **特征向量**，代表一个样本的所有特征
         
         * **多元线性回归模型向量化：**
           $$
           f_{\vec{w},b}(x) = \vec{w} \cdot \vec{x} + b
           $$
           *(注：点乘dot() ->元素级相乘，对应索引元素相乘再累加)*
      
   2. **多重线性回归的梯度下降：**
   
      1. **多特征成本函数：**
         $$
         J(\vec{w}, b) = \frac{1}{2m} \sum_{i=1}^{m} \left( f_{\vec{w},b}(\vec{x}^{(i)}) - y^{(i)} \right)^2
         $$
   
      2. **多元情况下参数更新：**
   
         * 参数w_n更新：
   
           对w的**偏导**
           $$
           \frac{\partial J}{\partial w_j} = \frac{1}{m} \sum_{i=1}^{m} \left( f_{\vec{w},b}(\vec{x}^{(i)}) - y^{(i)} \right) x_j^{(i)}
           $$
           参数w更新
           $$
           w_j = w_j - \alpha \frac{\partial}{\partial w_j} J(\vec{w}, b)
           $$
   
         * 参数b更新：
   
           对b的**偏导**
           $$
           \frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} \left( f_{\vec{w},b}(\vec{x}^{(i)}) - y^{(i)} \right)
           $$
           参数b更新
           $$
           b = b - \alpha \frac{\partial}{\partial b} J(\vec{w}, b)
           $$
   
   3. **特征缩放：**
   
      * 含义：把不同特征的数值范围调整到**相似的区间**
   
      * 意义：避免**成本函数**变成 “又窄又高” 的形状，导致梯度下降**收敛极慢**
   
        *(注：当特征尺度**差异很大**时，成本函数的等高线会成 **”细长型 “**或 **”瘦高型 “**，使梯度下降 ”之字型“ 导致**收敛极慢**)*
   
      * 常用方法：
   
        1. **Min-Max Scaling：**
           $$
           x_{\text{scaled}} = \frac{x - \min(x)}{\max(x) - \min(x)}
           $$
   
           * 区间：[0, 1]
   
           * 适用和特点：数据**分布**较**均匀**的数据，对**异常值**十分**敏感**
   
        2. **除以最大值：**（Min-Max Scaling简化版）
           $$
           x_{\text{scaled}} = \frac{x}{\max(x)}
           $$
   
           * 特点：数据分布可能**偏右**，最大值是1，最小值未必是0
   
        3. **Mean Normalization：**（均值归一化）
           $$
           x_{\text{scaled}} = \frac{x - \mu}{\max(x) - \min(x)}
           $$
   
           * 区间：大致[-1, 1]
   
        4. **Z-score Standardization：**（z分数归一化）
           $$
           x_{\text{scaled}} = \frac{x - \mu}{\sigma}
           $$
   
           * 特点：将所有数据缩放到**均值mean为0**，**标准差var为1**的正态分布，不受固定区间限制，更适合真是数据。对待异常值相对稳定
           
             *(注：公式中mean一般为**训练集均值**，std一般为**测试集标准差**)*
   
   
   
5. **逻辑回归模型：**

   1. **逻辑回归模型及其表达式：**

      1. 类别：分类模型（有监督的类别分类模型）

      2. **逻辑回归表达式：**（与Sigmoid函数复合）
         $$
         f_{\vec{w},b}(\vec{x}) = \frac{1}{1 + e^{-(\vec{w} \cdot \vec{x} + b)}}
         $$

         * 目的：使**y_hat**值对称**压缩到[-1, 1]上**，取逻辑真假值0与1（大于等于50%视为1，小于50%视为0）

   2.**成本函数——交叉熵函数：**（**Cross-Entropy**）

      1. **成本函数与损失函数：**

         * 单个样本的**损失函数**：
           $$
               L\left(f_{\vec{w},b}(\vec{x}^{(i)}), y^{(i)}\right) = \begin{cases} -\log\left(f_{\vec{w},b}(\vec{x}^{(i)})\right) & \text{if } y^{(i)} = 1 \\ -\log\left(1 - f_{\vec{w},b}(\vec{x}^{(i)})\right) & \text{if } y^{(i)} = 0 \end{cases}
           $$
         
           $$
           L\left(f_{\vec{w},b}(\vec{x}^{(i)}), y^{(i)}\right) = -y^{(i)} \log\left(f_{\vec{w},b}(\vec{x}^{(i)})\right) - (1 - y^{(i)}) \log\left(1 - f_{\vec{w},b}(\vec{x}^{(i)})\right)
           $$
         
           *(补：2式即1式的合并版)*
         
             * 全部样本的损失累加求均值：**成本函数：**
               $$
               = -\frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log\left(f_{\vec{w},b}(\vec{x}^{(i)})\right) + \left(1 - y^{(i)}\right) \log\left(1 - f_{\vec{w},b}(\vec{x}^{(i)})\right) \right]
               $$
   
   
   
   3.**参数更新（相同）：**
   
    * 参数w_n更新：
      
              对w的**偏导**
           $$
              \frac{\partial J}{\partial w_j} = \frac{1}{m} \sum_{i=1}^{m} \left( f_{\vec{w},b}(\vec{x}^{(i)}) - y^{(i)} \right) x_j^{(i)}
           $$
              参数w更新
           $$
              w_j = w_j - \alpha \frac{\partial}{\partial w_j} J(\vec{w}, b)
           $$
      
            * 参数b更新：
           
              对b的**偏导**
              $$
              \frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^{m} \left( f_{\vec{w},b}(\vec{x}^{(i)}) - y^{(i)} \right)
              $$
              参数b更新
              $$
              b = b - \alpha \frac{\partial}{\partial b} J(\vec{w}, b)
              $$
   





1. **K_means算法：**（无监督）

   1. 作用：用于**无监督学习**的进行**自行分类归类**的算法

   2. **原理**：

      * 随机挑选数据集中的**k个初始点**作为**初始质心**（簇的中心坐标）
      * 对每一个样本点计算其与个质心的距离（通常为**欧式距离**）
      * 对每个样本点逐一比较它们与各质心的距离，选取**距离最小**的质心作为其**归属簇**
      * 每轮迭代后，对簇内点计算其**均值点**即**新质心**，并将旧质心更新为新质心
      * 如此循环迭代，直到下一步的**质心位置不在变化**或**旧质心与新质心的位置距离小于给定误差**，即视为收敛，迭代结束
      * 如此，所有数据样本都将被分进一个簇堆，实现无监督数据的分类

   3. **优缺点：**

      * 优点：逻辑简单，代码实现容易

      * 缺点：

        1. 对于**初始k值选取**困难（需多实践测试选取最佳k值）

        2. 初始质心的位置对于最终收敛结果影响很大，即**初态对末态影响大**
        3. 复杂度与数据集容量线性增加，**数据量庞大则复杂度高**
        4. 对于特定图像，有一定几何形状的等特殊数据分布，该算法分类效果往往无法达到预期，即**对于特殊数据分布该算法效果不佳**

        

        
2. **PCA算法：**（无监督）

   1. **原理**：将原始的高维数据投影到**方差最大的方向**上，用更少的维度来保留数据的主要信息
      * 通俗而言：通过所谓**降维**，尽可能减少维度的同时保留尽可能多的数据信息，有**去噪**效果即舍去小方差成分，提高**信噪比**，同时降维**方便观察**并减少储存空间
   2. **优缺点**：
      * 优点：
        1. **无监督**，不需要标签适用于所有数据
        2. **线性特点**，计算简单
        3. **去噪**，舍弃小方差成分以**去除噪音**
      * 缺点：
        1. **方差与信息的不完全对应**，即方差大的不一定含信息量就大
        2. **对异常值敏感**，异常值对方差影响巨大
        3. **线性假设弊端**，只能捕捉线性相关性
   3. **维度灾难：**
      * 含义：随着**特征维度增加**，**数据空间**呈指数级增长，导致：
        - **数据稀疏**：样本在高维空间中变得极其稀疏
        - **计算爆炸**：计算距离的复杂度 O(n²) 到 O(n³)
        - **过拟合风险**：模型需要指数级更多的样本才能获得可靠的统计
      * PCA解法：
        1. ->数据稀疏：降维后样本在低维空间更密集，更**容易找到规律**
        2. ->计算复杂度高：从 d 维降到 k 维（k << d），**计算量大幅降低**
        3. ->过拟合：减少冗余特征，降低模型复杂度，**提高泛化能力**
        4. ->多重共线性：主成分之间相互正交（不相关），消除特征相关性
