#### **5 数据容器**

##### **1 列表list**

**(1)格式**：**列表名 = [v1, v2 ...]**

(注：列表元素可以为任意类型数据，也可以是另一个列表(嵌套))

(注：无元素即空列表，格式：`list1 = []` 或 `list2 = list()`)

**(注：列表元素 可以重复，可以修改)**

(注：列表可以不写列表名，直接写字面量)

(注：列表可以使用下标索引，**且负数意味着从尾部开始，即索引-1对应最后一个元素**)

**(2)列表的方法/函数**

* **index()函数：**

  格式：**列表名1.index(某元素)**

  作用：查找返回指定元素的下标索引，无则返回ValueError

* **insert()函数：**

  格式：**列表名1.insert(索引，元素值)**

  作用：=在指定索引位置，插入指定元素

* **append()函数：**

  格式：**列表名1.append(元素)**

  作用：将指定元素追加到列表尾部

* **extend()函数：**

  格式：**列表名1.extend(数据容器1)**

  作用：将指定数据容器1的所有元素取出，依次追加到列表尾部

* **pop()函数**

  格式：**列表名1.pop(索引)**

  作用：取出并返回对应索引的元素

  (注：取出后，原列表更新，被取出元素删除)

* **remove()函数**

  格式：**列表名1.remove(元素)**

  作用：删除指定元素的第一个匹配项

* **clear()函数**

  格式：**列表名1.clear()**

  作用：清空列表所有元素

* **count()函数**

  格式：**列表名1.count(元素)**

  作用：统计并返回对应元素在列表中的出现总数

* **len()函数**

  格式：**len(列表名1)**

  作用：统计并返回数据容器1的元素总个数

  *(注：已经不是列表专属)*

* **关键字del**

  格式：**del 列表名1[索引]**

  作用：删除指定索引的元素

  *(注：已经不是列表专属)*

**(2)元组的方法/函数**

* **index()函数：**

  格式：**元组名1.index(元素)**

* **count()函数：**

  格式：**元组名1.count(元素)**

* **len()函数：**

  格式：**len(元组名1)**



##### **3 字符串string**

**(1)格式：字符串名 = ”内容“**

(注：下标索引访问同列表)

**(注：元素为字符型，只读不改)**

**(2)字符串的方法/函数**

* **index()函数：**

  格式：**字符串名1.index(元素)**

* **count()函数：**

  格式：**字符串名1.count(元素)**

* **len()函数：**

  格式：**len(字符串名1)**

* **replace()函数：**

  格式：**字符串名1.replace(字符串a, 字符串b)**

  作用：将原字符串1副本中的字符串a替换为字符串2，然后返回

  **(注：不修改字符串1本身而是得一个新字符串)**

  (注：字符串a为字符串1的子串或本身)

* **split()函数：**

  格式：**字符串名1.split(分隔字符)**

  作用：按指定的分隔字符串，将原字符串1划分成多个新子字符串，并存入**列表**中，返回该列表

  (注：原字符串1不改变，得到一个新列表对象)

  (注：分隔字符串是原字符串1的一个子串)

  例：

  ```python
  str1 = "abcX1defX1gh"
  list1 = str1.split("X1")
  print(f"{list1}") #输出 ["abc", "def", "gh"]
  ```

* **strip()函数：**

  格式：**字符串名1.strip(指定字符串)**

  作用：去除前后指定的字符或字符集，指定字符串默认是**空白字符**

  (注：指定字符或指定字符集是**逐个匹配，不是整体匹配**)

  ```python
  s = "$$#hello#world$$#"
  result = s.strip("$#")  # 只要是$或#，首尾都会被删
  print(result)  # 输出：hello#world
  ```



##### **4 集合set**

**(1)格式：集合名1 = {v1, v2, ...}**

(注：元素可以为任意类型，亦可以是另一个集合(嵌套))

(注：无元素即空集合，格式：`set1 = set()`，**没有`set1 = {}`**的空集合写法，后者实为空字典)

**(注：集合元素 不可重复，可以修改)**

**(注：集合无序，不支持下标索引访问)**

**(2)集合的方法/函数**

* **remove()函数：**

  格式：**集合名.remove(元素)**

* **clear()函数：**

  格式：**集合名1.clear()**

* **pop()函数：**

  格式：**集合名1.pop()**

  作用：从集合中**随机抽取**一个元素出来并返回，原集合被修改

  (注：集合不支持下标索引，所以pop()中没有索引参数)

* **add()函数：**

  格式：**集合名1.add(元素)**

  作用：将指定元素加入到集合中

* **difference()函数：**

  格式：**集合名1.difference(集合名2)**

  作用：返回一个集合1与集合2的**差集**，差集为集合1中有而集合2中无的集合

* **difference_update()函数：**

  格式：**集合名1.difference_update(集合名2)**

  作用：仅仅改变集合1，将集合1内的与集合2相同的元素全部删除，集合2不改变

  (注：**没有返回值**，不返回新集合1，原集合1直接被改变)

* **union()函数：**

  格式：**集合名1.union(集合名2)**

  作用：返回集合1与集合2组合得到的新集合，两个原集合不变



##### **5 字典dict**

**(1)格式：字典名1 = {k1: v1, k2: v2, ...}**

(注：每一个元素都是一对**键值对**，**其中key唯一且不可为字典类型**，**value可以重复且可为字典(嵌套)**)

(注：无元素即空字典，格式：**`dict1 = {}`**或`dict1 = dict()`，**注意前者不是空集合写法**)

(注：字典不支持下标索引访问，访问格式：**字典名1[key] 去访问对应value**)

**(2)字典的方法/函数**

* **新增元素(访问)：**

  格式：**字典名[新key] = 新value**

  作用：添加新元素(键值对)到字典中

* **更新元素(访问)：**

  格式：**字典名[旧key] = 新value**

* **pop()函数：**

  格式：**字典名1.pop(key)**

  作用：返回对应key的value，原字典被修改

* **clear()函数：**

  格式：**字典名.clear()**

* **len()函数：**

  格式：**len(字典名1)**

* **keys()函数：**

  格式：**字典名.keys()**

  作用：返回一个包含所有key值的**列表**

  ```python
  dict1 = {'a': 100, 'b': 50, 'c': 0}
  for key1 in dict1:
      print(f"{key1}")
  for key2 in dict1.keys():
      print(f"{key2}")
  # 二者输出一样，后者显示写法，完全等价
  ```

**7 序列与切片**

**(1)序列：内容连续有序，支持下标索引的一类数据容器(如列表，元组，字符串...)**

**(2)切片：从序列中取出一个子序列的操作**

(3)格式：**序列名1[起始下标: 结束下标: 步长]**

(注：起始/结束下标可以省略，默认从开头元素(包含)起始，到末尾元素(不包含)结束，区间**左闭右开**)

(注：步长为摘取元素的间隔大小，省略默认为1，为负数则从末尾开始摘取)

(注：返回值(数据容器)类型与被切片的序列类型一致)
